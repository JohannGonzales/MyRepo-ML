```
#Vacios por columna
na_per_column = data[predictores_numericos].isna().sum()
na_per_column[na_per_column != 0].to_frame()
```

```
LotFrontage_na = data["LotFrontage"].notna()
GarageYrBlt_na = data["GarageYrBlt"].notna()
MasVnrArea_na = data["MasVnrArea"].notna()

mask_no_nas = np.logical_and(LotFrontage_na, GarageYrBlt_na)
mask_no_nas = np.logical_and(mask_no_nas, MasVnrArea_na)
del LotFrontage_na, GarageYrBlt_na , MasVnrArea_na
```

```
data_limpia = data.loc[mask_no_nas,:]
data_limpia["target"] = np.log10(data_limpia["SalePrice"])
```
